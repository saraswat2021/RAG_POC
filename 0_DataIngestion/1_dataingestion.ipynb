### DirectoryLoaders

```python
from langchain_community.document_loaders import DirectoryLoader
from langchain_community.document_loaders import TextLoader
# Load all text files from the directory
directoryLoader = DirectoryLoader(
    "data/files",
    glob="**/*.txt",
    loader_cls=TextLoader,
    # show_progress=True,
    loader_kwargs={"encoding": "utf-8"}
)
documents = directoryLoader.load()
print(f"Total documents loaded: {len(documents)}")
print(f"Document {documents}")
for i, doc in enumerate(documents):
    print(f"\nDocument {i+1}")
    print(f"Document source: {doc.metadata["source"][12:]}")
    print(f"Document content: {doc.page_content[:10]}...")  # Print
```

### Splitters for chunking

```python
from langchain_text_splitters import (
    CharacterTextSplitter,
    RecursiveCharacterTextSplitter,
    TokenTextSplitter
)
```

```python
text = documents[0].page_content
text
```

```python
char_splitter = CharacterTextSplitter(
    separator=" ", #split based on new line
    chunk_size=200, # size of each chunk
    chunk_overlap=50, # overlaping of characters in each chunk
    # length_function=len # to know length of function
)

splitted_chunks = char_splitter.split_text(text)
print(f"lenght of chunks : {len(splitted_chunks)}")
print(splitted_chunks)
for id, chunk in enumerate(splitted_chunks):
    print(f"chunk count : {id + 1}")
    print(f"chunk : {chunk}")
```

### Recursive Splitter

```python
chunk_splitter = RecursiveCharacterTextSplitter(
    separators=[" "],
    chunk_size=100,
    chunk_overlap=30,
    length_function=len
)

chunks = chunk_splitter.split_text(text)
print(f"length of chunks : {len(chunks)}")
for i in range(len(chunks)):
    print(f"chunk no : {i + 1}")
    print(chunks[i])
```

### Token Text Splitter

```python
tokenSplitter = TokenTextSplitter(
    chunk_size = 50,
    chunk_overlap = 10
)

chunks = tokenSplitter.split_text(text)
print(f"Total chunks {len(chunks)}")
for i, chunk in enumerate(chunks):
    print(f"Chunk no. : {i}")
    print(f"{chunk}")
    print("--------------------")

print("Note : Overlapping is necessary, suppose if two lines have some context with each other then model might pick all the connected lines")
```
