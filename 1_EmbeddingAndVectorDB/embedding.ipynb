```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model = "sentence-transformers/all-MiniLM-L6-v2")
```

```python
text = "Hello, I am learning RAG based systems"
embedding = embeddings.embed_query(text)
print(f"length of embedding {len(embedding)}")
```

```python
sentences = [
    "The cat sat on the mat",
    "The cat sat on the mat",
    "The dog played in the yard",
    "I love programming in Python",
    "Python is my favorite programming language"
]

embedding = embeddings.embed_documents(sentences)
print(f"length of embeddings {len(embedding)}")
print(embedding[0])
print(embedding[1])
```

```python
import numpy as np

def cosine_similarity(vector_1, vector_2):
    dot_product = np.dot(vector_1, vector_2)
    mod_1 = np.linalg.norm(vector_1)
    mod_2 = np.linalg.norm(vector_2)
    return (dot_product/(mod_1 * mod_2))

for i in range(len(sentences)):
    for j in range(i+1, len(sentences)):
        print(f"Similarity between sentences : {sentences[i]} V/s {sentences[j]}")
        similarity = cosine_similarity(embedding[i], embedding[j])
        print(f"Similarity is : {similarity}")
```
